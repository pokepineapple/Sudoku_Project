# NOTE

Applying Gaussian Blur on a said image will convert it from uint8 to a Float64 (CV_64F) and OpenCV's cv2.cvtColor function will not accept it.

```python
import imageio.v3 as iio
import matplotlib.pyplot as plt
import skimage as ski

import cv2
```

```python
image = iio.imread(uri=r"C:\Users\atcut\Documents\ML_Proj\Digits\9\img010-00005.png")
image = cv2.resize(image,(32,32))

# display the image
fig, ax = plt.subplots()
ax.imshow(image, cmap="gray")

image.dtype
```

```python
sigma = 2.5

# apply Gaussian blur, creating a new image
blurred = ski.filters.gaussian(
    image, sigma=(sigma, sigma), truncate=3.5, channel_axis=-1)
```

```python
print(image.ndim)
```

```python
# display blurred image
fig, ax = plt.subplots()
ax.imshow(blurred, cmap="gray")

blurred.dtype
```
